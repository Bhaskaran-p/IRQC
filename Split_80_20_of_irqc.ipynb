```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np

# --------------------------------------------------
# Load data
# --------------------------------------------------
movies = pd.read_csv('/content/movies.csv')
ratings = pd.read_csv('/content/ratings.csv')

# Merge datasets
df = movies.merge(ratings, how="left", on="movieId")
df.dropna(inplace=True)

print("Initial dataset shape:", df.shape)

# --------------------------------------------------
# Remove rare movies (SAFE threshold)
# --------------------------------------------------
comment_counts = df["title"].value_counts()

# Use smaller threshold to avoid empty dataset
threshold = 50
rare_movies = comment_counts[comment_counts <= threshold].index

df = df[~df["title"].isin(rare_movies)]

print("After removing rare movies:", df.shape)

# Safety check
if df.shape[0] == 0:
    print("Error: Dataset became empty after filtering. Reduce threshold.")
    exit()

# --------------------------------------------------
# Train-Test Split (80-20 BEFORE pivot)
# --------------------------------------------------
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

# --------------------------------------------------
# Create user-movie matrices
# --------------------------------------------------
train_user_movie_df = train_df.pivot_table(
    values="rating",
    index="userId",
    columns="title"
)

test_user_movie_df = test_df.pivot_table(
    values="rating",
    index="userId",
    columns="title"
)

# --------------------------------------------------
# Select movie
# --------------------------------------------------
movie_name = input("Enter movie name: ")

if movie_name not in train_user_movie_df.columns:
    print("Error: Movie not found in training data.")
    exit()

# --------------------------------------------------
# Compute correlations (TRAIN data only)
# --------------------------------------------------
movie_name_df = train_user_movie_df[movie_name]

movie_corr = train_user_movie_df.corrwith(movie_name_df).sort_values(ascending=False)
movie_corr = movie_corr.dropna()

# Remove self-correlation
movie_corr = movie_corr.drop(movie_name, errors='ignore')

# Top 5 recommendations
top_correlated_movies = movie_corr.head(5)

print(f"\nTop 5 movies similar to '{movie_name}':")
print(top_correlated_movies)


```

```python
import pennylane as qml
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Load and preprocess data
df = pd.read_csv('/content/BigBasket Products.csv')
df.dropna(inplace=True)

# =========================
# 80-20 SPLIT
# =========================
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Pivot tables
train_user_product_df = train_df.pivot_table("sale_price", "rating", "product")
test_user_product_df = test_df.pivot_table("sale_price", "rating", "product")

# Align columns (VERY IMPORTANT)
common_products = train_user_product_df.columns.intersection(test_user_product_df.columns)
train_user_product_df = train_user_product_df[common_products]
test_user_product_df = test_user_product_df[common_products]

# =========================
# QUANTUM CIRCUIT
# =========================
n_qubits = 5
dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev)
def quantum_corr(params):

    qml.templates.AngleEmbedding(params[:n_qubits], wires=range(n_qubits), rotation="X")

    for i in range(n_qubits):
        qml.Hadamard(wires=i)

    for i in range(n_qubits):
        qml.RX(params[i], wires=i)
        qml.RY(params[i + n_qubits], wires=i)

    for i in range(n_qubits - 1):
        qml.CZ(wires=[i, i + 1])

    for i in range(n_qubits - 1):
        qml.CNOT(wires=[i, i + 1])

    return qml.expval(qml.prod(*[qml.PauliZ(i) for i in range(n_qubits)]))

# =========================
# INPUT PRODUCT
# =========================
product_name = input("Enter Product name: ")

if product_name not in train_user_product_df.columns:
    raise ValueError("Product not found in TRAIN data")

# =========================
# GROUND TRUTH FROM TEST
# =========================
ground_truth_matrix = test_user_product_df.corr().fillna(0)

# =========================
# QUANTUM CORRELATION FROM TRAIN
# =========================
product_correlations = []

for column in train_user_product_df.columns:
    if column != product_name:
        params = np.random.uniform(0, 2 * np.pi, size=2 * n_qubits)
        correlation = quantum_corr(params)
        product_correlations.append((column, correlation))

# Sort top-5
product_correlations.sort(key=lambda x: x[1], reverse=True)
top_correlated = product_correlations[:5]

print("\nTop correlated products:")
for product, correlation in top_correlated:
    print(f"{product}: {correlation:.4f}")

# =========================
# METRICS
# =========================
quantum_correlations = []
ground_truth_correlations = []

for product, corr in top_correlated:
    if product in ground_truth_matrix.columns:
        quantum_correlations.append(corr)
        ground_truth_correlations.append(ground_truth_matrix.loc[product_name, product])

mae = mean_absolute_error(quantum_correlations, ground_truth_correlations)
mse = mean_squared_error(quantum_correlations, ground_truth_correlations)
rmse = np.sqrt(mse)

print(f"\nMAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
```

```python
import pennylane as qml
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

# --------------------------------------------------
# Load and preprocess data
# --------------------------------------------------
movies = pd.read_csv('/content/movies.csv')
ratings = pd.read_csv('/content/ratings.csv')

df = movies.merge(ratings, how="inner", on="movieId")
df.dropna(inplace=True)

# --------------------------------------------------
# Train-Test Split (80-20 BEFORE pivot)
# --------------------------------------------------
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Pivot tables
train_user_movie_df = train_df.pivot_table(
    values="rating",
    index="userId",
    columns="title"
).fillna(0)

test_user_movie_df = test_df.pivot_table(
    values="rating",
    index="userId",
    columns="title"
).fillna(0)

# --------------------------------------------------
# Quantum setup
# --------------------------------------------------
n_qubits = 5
dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev)
def quantum_corr(params):
    for i in range(n_qubits):
        qml.Hadamard(wires=i)

    for i in range(n_qubits):
        qml.RX(params[i], wires=i)
        qml.RY(params[i + n_qubits], wires=i)

    for i in range(n_qubits - 1):
        qml.CZ(wires=[i, i + 1])
        qml.CNOT(wires=[i, i + 1])

    return qml.expval(qml.prod(*[qml.PauliZ(i) for i in range(n_qubits)]))

# --------------------------------------------------
# Metrics
# --------------------------------------------------
def recall_at_k(predicted, ground_truth, k, threshold=0.1):
    relevant_items = np.where(ground_truth > threshold)[0]
    if len(relevant_items) == 0:
        return 0.0

    top_k_items = np.argsort(predicted)[::-1][:k]
    hits = len(set(top_k_items) & set(relevant_items))
    return hits / len(relevant_items)


def ndcg_at_k(predicted, ground_truth, k):
    order = np.argsort(predicted)[::-1][:k]
    gains = ground_truth[order]

    discounts = 1 / np.log2(np.arange(2, k + 2))
    dcg = np.sum(gains * discounts)

    ideal_order = np.argsort(ground_truth)[::-1][:k]
    ideal_gains = ground_truth[ideal_order]
    idcg = np.sum(ideal_gains * discounts)

    return dcg / idcg if idcg > 0 else 0.0

# --------------------------------------------------
# Select movie
# --------------------------------------------------
movie_name = input("Enter movie name: ")

if movie_name not in train_user_movie_df.columns:
    raise ValueError("Movie not found in TRAIN data")

if movie_name not in test_user_movie_df.columns:
    raise ValueError("Movie not found in TEST data")

# --------------------------------------------------
# Ground truth from TEST set
# --------------------------------------------------
ground_truth_matrix = test_user_movie_df.corr().fillna(0)

# --------------------------------------------------
# Quantum predictions (TRAIN set)
# --------------------------------------------------
predicted_scores = []
true_scores = []
movie_list = []

for movie in train_user_movie_df.columns:
    if movie != movie_name and movie in ground_truth_matrix.columns:

        params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)
        q_score = quantum_corr(params)

        gt_score = ground_truth_matrix.loc[movie_name, movie]

        predicted_scores.append(q_score)
        true_scores.append(gt_score)
        movie_list.append(movie)

predicted_scores = np.array(predicted_scores)
true_scores = np.array(true_scores)

# --------------------------------------------------
# Top-5 Recommendations
# --------------------------------------------------
top5_idx = np.argsort(predicted_scores)[::-1][:5]

print("\nTop-5 Quantum Recommendations:")
for i in top5_idx:
    print(f"{movie_list[i]} : {predicted_scores[i]:.4f}")

# --------------------------------------------------
# Error Metrics
# --------------------------------------------------
mae = mean_absolute_error(true_scores, predicted_scores)
mse = mean_squared_error(true_scores, predicted_scores)
rmse = np.sqrt(mse)

print("\nError Metrics:")
print(f"MAE  : {mae:.4f}")
print(f"MSE  : {mse:.4f}")
print(f"RMSE : {rmse:.4f}")

# --------------------------------------------------
# Ranking Metrics (K = 2,3,4,5)
# --------------------------------------------------
print("\nRanking Metrics:")
for k in [2, 3, 4, 5]:
    rec = recall_at_k(predicted_scores, true_scores, k)
    ndcg = ndcg_at_k(predicted_scores, true_scores, k)

    print(f"K={k} | Recall@{k}: {rec:.4f} | NDCG@{k}: {ndcg:.4f}")
```

```python
import pennylane as qml
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# --------------------------------------------------
# Load and preprocess data
# --------------------------------------------------
df = pd.read_csv('/content/supermarket_sales - Sheet1.csv')
df.dropna(inplace=True)

# --------------------------------------------------
# Train-Test Split (80-20)
# --------------------------------------------------
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Pivot tables
train_user_product_df = train_df.pivot_table(
    values="Unit price",
    index="Rating",
    columns="Product line"
)

test_user_product_df = test_df.pivot_table(
    values="Unit price",
    index="Rating",
    columns="Product line"
)

# Ground truth from TEST set
ground_truth_matrix = test_user_product_df.corr().fillna(0)

# --------------------------------------------------
# Quantum circuit setup
# --------------------------------------------------
n_qubits = 5
dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev)
def quantum_corr(params):
    # Angle embedding
    qml.templates.AngleEmbedding(
        params[:n_qubits], wires=range(n_qubits), rotation="X"
    )

    # Hadamard gates
    for i in range(n_qubits):
        qml.Hadamard(wires=i)

    # RX and RY rotations
    for i in range(n_qubits):
        qml.RX(params[i], wires=i)
        qml.RY(params[i + n_qubits], wires=i)

    # Entanglement
    for i in range(n_qubits - 1):
        qml.CZ(wires=[i, i + 1])
        qml.CNOT(wires=[i, i + 1])

    return qml.expval(qml.prod(*[qml.PauliZ(i) for i in range(n_qubits)]))

# --------------------------------------------------
# Ranking metrics
# --------------------------------------------------
def recall_at_k(quantum_ranking, ground_truth_ranking, k):
    q_top = set(quantum_ranking[:k])
    gt_top = set(ground_truth_ranking[:k])

    if len(gt_top) == 0:
        return 0.0

    return len(q_top & gt_top) / len(gt_top)


def ndcg_at_k(quantum_ranking, ground_truth_scores, k):
    def dcg(scores):
        return np.sum([
            (2**rel - 1) / np.log2(i + 2)
            for i, rel in enumerate(scores)
        ])

    rel_scores = [
        ground_truth_scores.get(prod, 0.0)
        for prod in quantum_ranking[:k]
    ]

    ideal_scores = sorted(
        ground_truth_scores.values(), reverse=True
    )[:k]

    idcg = dcg(ideal_scores)
    if idcg == 0:
        return 0.0

    return dcg(rel_scores) / idcg

# --------------------------------------------------
# Select target product
# --------------------------------------------------
product_name = input("Enter Product name: ")

if product_name not in train_user_product_df.columns:
    print("Error: Product not found in training data.")
    exit()

# --------------------------------------------------
# Compute quantum correlations (TRAIN data)
# --------------------------------------------------
product_correlations = []

for column in train_user_product_df.columns:
    if column != product_name:
        params = np.random.uniform(0, 2 * np.pi, size=2 * n_qubits)
        corr_value = quantum_corr(params)
        product_correlations.append((column, corr_value))

# Sort by quantum correlation
product_correlations.sort(key=lambda x: x[1], reverse=True)
quantum_ranking = [prod for prod, _ in product_correlations]

# --------------------------------------------------
# Ground-truth ranking (TEST data)
# --------------------------------------------------
if product_name not in ground_truth_matrix.columns:
    print("Warning: Product not found in test set.")
    exit()

ground_truth_ranking = (
    ground_truth_matrix[product_name]
    .drop(product_name)
    .sort_values(ascending=False)
    .index.tolist()
)

ground_truth_scores = (
    ground_truth_matrix[product_name]
    .drop(product_name)
    .to_dict()
)

# --------------------------------------------------
# Evaluation for multiple K values
# --------------------------------------------------
K_values = [2, 3, 4, 5]

print("\nTop Quantum-Correlated Products:")
for prod, score in product_correlations[:5]:
    print(f"{prod}: {score:.4f}")

print("\nEvaluation Metrics:")
print("K\tRecall@K\tNDCG@K")

for K in K_values:
    recall_k = recall_at_k(quantum_ranking, ground_truth_ranking, K)
    ndcg_k = ndcg_at_k(quantum_ranking, ground_truth_scores, K)

    print(f"{K}\t{recall_k:.4f}\t\t{ndcg_k:.4f}")
```
