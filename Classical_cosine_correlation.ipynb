MovieLens 10k

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load movies and ratings data
movies = pd.read_csv('/content/movies.csv')
ratings = pd.read_csv('/content/ratings.csv')

# Merge the data
df = movies.merge(ratings, how="left", on="movieId")
df.dropna(inplace=True)

# Count the number of ratings for each movie
comment_counts = pd.DataFrame(df["title"].value_counts(), columns=["count"])

# Filter out movies with less than 50 ratings
rare_movies = comment_counts[comment_counts["count"] <= 50].index
common_movies = df[~df["title"].isin(rare_movies)]

# Create user-movie matrix
user_movie_df = common_movies.pivot_table(
    values="rating",
    index="userId",
    columns="title"
)

# Fill NaN with 0 for cosine similarity
user_movie_df_filled = user_movie_df.fillna(0)

# Train-test split (not strictly needed for similarity, kept as in your code)
train, test = train_test_split(user_movie_df_filled, test_size=0.2, random_state=42)

# Select a movie
movie_name = input("Enter movie name: ")

if movie_name in user_movie_df_filled.columns:
    # Extract the selected movie vector
    movie_vector = user_movie_df_filled[movie_name].values.reshape(1, -1)

    # Compute cosine similarity with all movies
    similarity_scores = cosine_similarity(
        movie_vector,
        user_movie_df_filled.T.values
    ).flatten()

    # Create similarity dataframe
    movie_similarity = pd.Series(
        similarity_scores,
        index=user_movie_df_filled.columns
    ).sort_values(ascending=False)

    # Remove the input movie itself
    movie_similarity = movie_similarity.drop(movie_name)

    # Top 5 similar movies
    top_similar_movies = movie_similarity.head(5)

    print(f"\nTop 5 movies similar to '{movie_name}' (Cosine Similarity):")
    print(top_similar_movies)

```

#IMDB Top 250 Movies

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Read the CSV file
df = pd.read_csv('/content/IMDB Top 250 Movies.csv')

# Drop rows with missing values
df.dropna(inplace=True)

# Create a pivot table
# Rows: rating values, Columns: movie names, Values: year (as in your original code)
user_movie_df = df.pivot_table(
    values="year",
    index="rating",
    columns="name"
)

# Fill missing values with 0 (required for cosine similarity)
user_movie_df.fillna(0, inplace=True)

# Input movie name
movie_name = input("Enter movie name: ")

if movie_name in user_movie_df.columns:
    # Selected movie vector
    movie_vector = user_movie_df[movie_name].values.reshape(1, -1)

    # Compute cosine similarity with all movies
    cosine_scores = cosine_similarity(
        movie_vector,
        user_movie_df.T.values
    ).flatten()

    # Create similarity series
    movie_sim = pd.Series(
        cosine_scores,
        index=user_movie_df.columns
    ).sort_values(ascending=False)

    # Remove the input movie itself
    movie_sim = movie_sim.drop(movie_name)

    # Top 5 similar movies
    top_5_sim_movies = movie_sim.head(5)
    print("\nTop 5 similar movies (Cosine Similarity):")
    print(top_5_sim_movies)

```
